```python
!pip install yfinance==0.2.4
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


```

```python
tesla = yf.Ticker("TSLA")
```

```python

tesla_data = tesla.history(period="max")

# Imprimir los primeros registros
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python

tesla_data.reset_index(inplace=True, drop=True)
plt.figure(figsize=(10, 6))
plt.plot(tesla_data['Date'], tesla_data['Open'], label='Open Price', color='red')
plt.title('Tesla Stock Price over the years')
plt.xlabel('Date')
plt.ylabel('Open price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from bs4 import BeautifulSoup 
import requests  
```

```python
url = "https://companiesmarketcap.com/tesla/revenue/"
html_data = requests.get(url).text
```

```python
soup = BeautifulSoup(html_data,"html5lib")
```

```python
tesla_revenue = pd.DataFrame(columns=["Year",  "Revenue"])
```

```python
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    year = col[0].text
    revenue = col[1].text
    
    tesla_revenue = tesla_revenue.append({"Year":year, "Revenue":revenue }, ignore_index=True)    
```

```python
tesla_revenue.head()
```

```python
tesla_revenue.tail()
```
